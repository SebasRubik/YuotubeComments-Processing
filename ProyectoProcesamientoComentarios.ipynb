¿Qué opiniones expresan los usuarios en los comentarios de los videos del canal de YouTube miduDev sobre la creciente presencia de programadores junior en la industria del software?

```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import pandas as pd

# Descargar los recursos necesarios de nltk
nltk.download('punkt')
nltk.download('stopwords')
nltk.download('punkt_tab')
```

```python
# Inicializar el Stemmer y las Stopwords
stemmer = PorterStemmer()
stop_words = set(stopwords.words('spanish'))  # Se Puede cambiar el idioma si los comentarios están en español

# Función de procesamiento de texto
def procesar_comentario(texto):
    # Tokenización
    tokens = word_tokenize(texto.lower())  # Convertir a minúsculas y tokenizar
    # Eliminar stopwords y aplicar stemming
    tokens_procesados = [stemmer.stem(palabra) for palabra in tokens if palabra.isalnum() and palabra not in stop_words]
    # Unir tokens procesados de nuevo en una cadena
    return ' '.join(tokens_procesados)

# Aplicar el procesamiento a la columna de comentarios
df['comentarios_procesados'] = df['text'].apply(procesar_comentario)

# Mostrar el DataFrame con los comentarios procesados
df[['text', 'comentarios_procesados']].head()
```

```python
import nltk
nltk.download('vader_lexicon')
```

```python
from nltk.sentiment import SentimentIntensityAnalyzer
# Inicializar el analizador de sentimientos
sia = SentimentIntensityAnalyzer()

# Función para analizar los sentimientos
def obtener_sentimiento(comentario):
    return sia.polarity_scores(comentario)

# Aplicar el análisis de sentimientos a cada comentario
df['sentimientos'] = df['comentarios_procesados'].apply(obtener_sentimiento)

# Mostrar los resultados
df[['comentarios_procesados', 'sentimientos']].head()
```

```python
# Extraer el puntaje compuesto del análisis de sentimientos
df['sentimiento_compuesto'] = df['sentimientos'].apply(lambda x: x['compound'])
# Función para clasificar el sentimiento
def clasificar_sentimiento(score):
    if score >= 0.05:
        return 'Positivo'
    elif score <= -0.05:
        return 'Negativo'
    else:
        return 'Neutral'

# Aplicar la clasificación de sentimiento
df['categoria_sentimiento'] = df['sentimiento_compuesto'].apply(clasificar_sentimiento)

# Mostrar los resultados con la categoría de sentimiento
df[['comentarios_procesados', 'sentimiento_compuesto', 'categoria_sentimiento']]
```

```python
neutral_count = (df['categoria_sentimiento'] == 'Neutral').sum()
negative_count = (df['categoria_sentimiento'] == 'Negativo').sum()
postive_count = (df['categoria_sentimiento'] == 'Positivo').sum()

print('Comentarios neutrales: ',neutral_count)
print('Comentarios negativos',negative_count)
print('Comentarios positivos',postive_count)
```

```python
df = df.drop(columns=['Unnamed: 0','cid','text','time','author','channel','votes','photo','heart','reply','time_parsed'])
df.head()
```

```python
# Topic Modeling
import torch.nn.functional as F

from torch import Tensor
from transformers import AutoTokenizer, AutoModel
```

```python
tokenizer = AutoTokenizer.from_pretrained('intfloat/multilingual-e5-small')
model = AutoModel.from_pretrained('intfloat/multilingual-e5-small')
```

```python
def average_pool(last_hidden_states: Tensor,
                 attention_mask: Tensor) -> Tensor:
    last_hidden = last_hidden_states.masked_fill(~attention_mask[..., None].bool(), 0.0)
    return last_hidden.sum(dim=1) / attention_mask.sum(dim=1)[..., None]
```

```python
def toEmbbeding(comment):
    tokenized = tokenizer(comment,padding=True, truncation=True, return_tensors='pt')
    outputs = model(**tokenized)
    embeddings = average_pool(outputs.last_hidden_state, tokenized['attention_mask'])
    embeddings = F.normalize(embeddings, p=2, dim=1)
    list_embedding = embeddings.cpu().detach().numpy().tolist()[0]
    return list_embedding
```

```python
dfEmbbeded = df
dfEmbbeded.head()
```

```python
dfEmbbeded['Embeddings'] = None
for index , fila in dfEmbbeded.iterrows():
    dfEmbbeded.at[index,'Embeddings'] = toEmbbeding(fila['comentarios_procesados'])

dfEmbbeded.head()
```

```python
df['Embeddings'].head()
```

```python
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans


inertia = []
k_values = range(1,20)  # Probar diferentes valores de k
for k in k_values:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(list(dfEmbbeded['Embeddings']))
    inertia.append(kmeans.inertia_)

# Graficar el Método del Codo
plt.plot(k_values, inertia, 'bo-')
plt.xlabel('Número de Clusters (K)')
plt.ylabel('Inercia')
plt.title('Método del Codo para elegir el número óptimo de clusters')
plt.show()
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 3)
kmeans.fit(list(dfEmbbeded['Embeddings']))
labels = kmeans.labels_
```

```python
dfEmbbeded['Cluster'] = labels
```
